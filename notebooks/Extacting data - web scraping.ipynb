```python
import requests
from bs4 import BeautifulSoup
```

```python
from IPython.core.display import display, HTML
```

```python
result = requests.get('https://www.ynet.co.il')
```

```python
result.status_code
```

```python
display(HTML(result.text))
```

```python
ps = BeautifulSoup(result.text)
```

```python
print(ps)
```

```python
print(ps.find(name="body"))
```

```python
head = ps.find(name="head")
print(head)
```

```python
title = head.find(name='title')
print(title)
```

```python
links = ps.findAll(name='a')
print(links)
```

```python
for p in ps.findAll(name='p', attrs={"class" : "more_news"}):
    print(p)
```
